```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t_array = np.linspace(0, 20, 1000)
```

```python
def possibility_one(A, B, c, m, d, t):
    return A*np.exp(-c*t/(2*m))*np.cos(d*t/(2*m)) + B*np.exp(-c*t/(2*m))*np.sin(d*t/(2*m))

def possibility_two(A, B, c, m, d, t):
    return A*np.exp(-c*t/(2*m)) + B*t*np.exp(-c*t/(2*m))

def possibility_three(A, B, c, m, d, t):
    return A*np.exp(-(c+d)*t/(2*m)) + B*np.exp(-(c-d)*t/(2*m))
```

```python
c = 1
m = 1
k = 4
d = np.sqrt(4*m*k - c**2)
A = 1 
B = c/d
```

```python
fig, ax = plt.subplots()
ax.plot(t_array, possibility_one(1, 1/np.sqrt(4*m*k - c**2), 1, 1, np.sqrt(4*1*4 - 1**2), t_array), label='Possibility One')
ax.plot(t_array, possibility_two(1, 1/np.sqrt(4*m*k - c**2), 1, 1, np.sqrt(4*1*4 - 1**2), t_array), label='Possibility Two')
ax.plot(t_array, possibility_three(1, -0.5, 5, 1, np.sqrt(5**2 - 4*1*4), t_array), label='Possibility Three')
plt.legend()
fig.show()
```
